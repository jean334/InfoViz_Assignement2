```python
#order by year
df = df.sort_values(by='Year')
df['Platform']
```

```python
bar_dict = []
for elem in ["Instagram", "Twitter", "Facebook"]:
    bar_dict.append({
        "social media": elem,
        "positive": np.zeros(2023-2010+1),
        "neutral": np.zeros(2023-2010+1),
        "negative": np.zeros(2023-2010+1)
    })

social_media_mapping = {"Instagram":0, "Twitter":1, "Facebook":2}

for i in range(len(df)):
    entry = df.iloc[i]
    if entry['Platform'] == 'Instagram':
        index = 0
    elif entry['Platform'] == 'Twitter':
        index = 1
    elif entry['Platform'] == 'Facebook':
        index = 2
    

    if entry['sentiment'] == 'positive':
        bar_dict[index]['positive'][entry['Year']-2010] += 1
    elif entry['sentiment'] == 'neutral':
        bar_dict[index]['neutral'][entry['Year']-2010] += 1
    else:
        bar_dict[index]['negative'][entry['Year']-2010] += 1

bar_dict
```

```python
import copy
bar_dict_percent = copy.deepcopy(bar_dict)
bar_dict_percent
```

```python
def get_percentages(bar_dict):
    for i in range(len(bar_dict)):
        total = bar_dict[i]['positive'] + bar_dict[i]['neutral'] + bar_dict[i]['negative']

        bar_dict_percent[i]['positive'] = (bar_dict[i]['positive'] / total) * 100
        bar_dict_percent[i]['neutral'] = (bar_dict[i]['neutral'] / total) * 100
        bar_dict_percent[i]['negative'] = (bar_dict[i]['negative'] / total) * 100

get_percentages(bar_dict_percent)
bar_dict_percent

def replaceNaNbyzero(bar_dict_percent):
    for i in range(len(bar_dict_percent)):
        for key in bar_dict_percent[i]:
            bar_dict_percent[i][key] = np.nan_to_num(bar_dict_percent[i][key])
    return bar_dict_percent

bar_dict_percent = replaceNaNbyzero(bar_dict_percent)
```
