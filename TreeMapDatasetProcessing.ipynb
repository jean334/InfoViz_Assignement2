```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Dataset/train.csv')
df
```

```python
def createBranches():
    branches = []
    for i in range(len(df)):
        branches.append([df.loc[i]['Platform'], df.loc[i]['Dominant_Emotion']])
    return branches

branches = createBranches()
```

```python
branches
```

```python
def insert_branch(root, branch):
    """ Insert a branch into a tree """
    node = root
    for i, name in enumerate(branch):
        if "children" in node:
            found = next((child for child in node["children"] if child["name"] == name), None)
        
        if found:
            print("found")
            node = found
            print(node)
            if i == len(branch) - 1:
                node["value"] += 1
            """
            for j in range(len(node["children"])):
                if node["children"][j]["name"] == name:
                    node["children"][j]["value"] += 1
            """
        
        else:
            new_node = {"name": name}
            if i == len(branch) - 1:
                new_node["value"] = 1  
            else:
                new_node["children"] = []
            node["children"].append(new_node)
            node = new_node

def build_tree(branches):
    """ Build a tree from a list of branches """
    root = {"name": "root", "children": []} 
    for branch in branches:
        insert_branch(root, branch)
    return root
```

```python
import json
tree = build_tree(branches)

if len(tree["children"]) == 1:
    tree = tree["children"][0]

with open("./Dataset/treeMap.json", "w", encoding="utf-8") as f:
    json.dump(tree, f, indent=4)

print(json.dumps(tree, indent=4))
```
